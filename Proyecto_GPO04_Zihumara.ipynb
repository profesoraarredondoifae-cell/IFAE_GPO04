# **Interpretación de los resultados**

El análisis de la producción mensual revela picos y caídas significativas que requieren una interpretación más profunda. Observamos **disminuciones notables en la producción durante los meses de Febrero, Julio y Diciembre**. Estos periodos coinciden directamente con la alta rotación de personal mencionada.

### Observaciones Clave:

1.  **Rotación de Personal y Producción:** Los gráficos de barras y de tendencia lineal muestran claramente una baja en la producción en Febrero (4.4% del total anual), Julio (6.6%) y Diciembre (5.0%). Esta correlación sugiere que la inestabilidad del personal en estos meses afecta directamente la capacidad productiva.

2.  **Impacto de la Ausencia de Capacitación:** La falta de un plan de capacitación formal agrava esta situación. Un nuevo personal sin la formación adecuada o el personal existente que no recibe actualizaciones continuas puede llevar a ineficiencias, errores y, en última instancia, a una menor producción.

### Propuesta: Plan de Capacitación y Crecimiento Teórico

La implementación de un **plan de capacitación robusto** es crucial para mitigar los efectos de la rotación de personal y mejorar la eficiencia general. Teóricamente, con un programa de capacitación efectivo, podríamos esperar el siguiente escenario de crecimiento:

*   **Estabilización en Febrero, Julio y Diciembre:** Al reducir la curva de aprendizaje y mejorar la retención de personal capacitado, estos meses podrían ver un aumento de producción que los alinee más con los meses de rendimiento promedio (aproximadamente 8-9% del total anual por mes).
*   **Mejora General de la Eficiencia:** La capacitación continua no solo beneficia a los nuevos empleados, sino que también mejora las habilidades y la productividad del personal existente, lo que podría resultar en un **incremento teórico de la producción total anual en un 10-15%**.
*   **Reducción de "Scrap" y Errores:** Un personal bien capacitado comete menos errores, lo que se traduce en una disminución de piezas de descarte (scrap) y un mejor aprovechamiento de los recursos.

**Panorama de Crecimiento Teórico:**

Si consideramos un aumento del 10% en la producción de los meses históricamente bajos (Febrero, Julio, Diciembre) y un 5% en los demás meses debido a la mejora general de la eficiencia, podríamos proyectar una producción anual significativamente mayor y más estable. Este crecimiento teórico representa un argumento sólido para la inversión en capacitación como estrategia clave para la mejora continua y la optimización de la producción.

# **Predicción**

```python
# @title
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# --- 1. Prepare data for 2026 prediction (using Jan-Apr 2026 trend) ---
# We use only the first 4 months of 2026 from monthly_df_2026 for the regression
# to capture the initial strong positive trend before the May dip.
# Reshape X for sklearn: (-1, 1) means 1 column, as many rows as needed.
X_train_2026 = monthly_df_2026['month_num'].iloc[:4].values.reshape(-1, 1)
y_train_2026 = monthly_df_2026['produccion'].iloc[:4].values

# --- 2. Fit Linear Regression Model ---
model_2026 = LinearRegression()
model_2026.fit(X_train_2026, y_train_2026)

# --- 3. Predict for remaining months (May-December) based on the Jan-Apr trend ---
# Predict for months 5 through 12
X_predict_months = np.array(range(5, 13)).reshape(-1, 1)
y_predict_2026_future = model_2026.predict(X_predict_months)

# Ensure predicted values are non-negative
y_predict_2026_future[y_predict_2026_future < 0] = 0

# --- 4. Combine actual and predicted 2026 data ---
# Start with actual data for months 1-4
monthly_df_2026_combined = monthly_df_2026.iloc[:4].copy()

# Create a DataFrame for predicted months 5-12
predicted_df = pd.DataFrame({
    'month_num': range(5, 13),
    'produccion': y_predict_2026_future.astype(int)
})

# Map month numbers to Spanish names for predicted months
spanish_month_names_list = [
    'Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
    'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'
]
predicted_df['month_name'] = predicted_df['month_num'].apply(lambda x: spanish_month_names_list[x-1])

# Append predicted data to the combined DataFrame
monthly_df_2026_combined = pd.concat([monthly_df_2026_combined, predicted_df], ignore_index=True)

# --- Apply a boosting factor to ensure theoretical elevation for predicted months ---
# This ensures that the 'elevated' aspect of the theoretical projection is met.
# For months 5-12, ensure predicted production is at least 15% higher than the corresponding 2025 month.
for i in range(len(monthly_df_2026_combined)):
    month = monthly_df_2026_combined.loc[i, 'month_num']
    if month > 4: # Only for predicted months (May-December)
        # Get the corresponding month's production from 2025
        prod_2025_month = monthly_df[monthly_df['month_num'] == month]['produccion'].iloc[0]
        # Set a minimum predicted value for 2026 to be 15% higher than 2025's corresponding month
        target_min_2026 = prod_2025_month * 1.15
        if monthly_df_2026_combined.loc[i, 'produccion'] < target_min_2026:
            monthly_df_2026_combined.loc[i, 'produccion'] = int(target_min_2026)

# --- 5. Create the comparative bar chart ---
plt.figure(figsize=(16, 8))

bar_width = 0.35
index = np.arange(len(spanish_month_names_list)) # 12 months for x-axis

# Bars for 2025 production
plt.bar(index - bar_width/2, monthly_df['produccion'], bar_width, label='Producción 2025 (Real)', color='skyblue')

# Bars for 2026 production (actual + predicted)
plt.bar(index + bar_width/2, monthly_df_2026_combined['produccion'], bar_width, label='Producción 2026 (Real/Proyectado)', color='lightcoral')

# Add titles and labels with larger, bold font
plt.title('Comparativa de Producción Mensual 2025 vs. 2026 (Proyectado)', fontsize=24, fontweight='bold')
plt.xlabel('Mes', fontsize=20, fontweight='bold')
plt.ylabel('Producción Total', fontsize=20, fontweight='bold')
plt.xticks(index, spanish_month_names_list, rotation=45, ha='right', fontsize=16, fontweight='bold')
plt.yticks(fontsize=16, fontweight='bold')
plt.legend(fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust layout to prevent labels from overlapping

plt.show()

# Display the projected 2026 annual total and month-by-month values.
total_produccion_anual_2026_projected = monthly_df_2026_combined['produccion'].sum()
print(f"\nProducción Total Anual Proyectada para 2026: {total_produccion_anual_2026_projected:,} Unidades")
print("\nProducción Mensual Detallada para 2026 (Real/Proyectado):")
display(monthly_df_2026_combined)
```

# **Proyección de los resultados**

La proyección tan positiva para el año 2026, donde la producción teórica se eleva significativamente en comparación con 2025, se fundamenta en la aplicación de un modelo de mejora continua enfocado en el factor humano, que he inferido como FAI (Formación, Adaptación e Incentivo). Este modelo busca abordar directamente las causas de la variabilidad y las bajas en la producción observadas en 2025 (como la alta rotación y la falta de capacitación).

Aquí te explico cómo los principios del modelo FAI contribuyen a esta proyección positiva:

Formación (Capacitación Continua y Especializada):

Reducción de la Curva de Aprendizaje: El plan de capacitación implementado asegura que el nuevo personal alcance rápidamente los niveles de productividad óptimos, minimizando el impacto de la rotación. Como se mencionó en el análisis de 2025, la falta de capacitación agravaba la situación de rotación.
Mejora de la Eficiencia: El personal existente recibe actualizaciones continuas y formación en nuevas técnicas, lo que se traduce en mayor eficiencia, menos errores y un incremento general en la velocidad de producción. Un personal bien capacitado comete menos errores, lo que lleva a una reducción del "scrap" (piezas de descarte) y un mejor aprovechamiento de los recursos.
Versatilidad: La capacitación multifuncional permite que los empleados puedan desempeñarse en diferentes áreas, reduciendo cuellos de botella y mejorando la flexibilidad operativa, especialmente en picos de demanda o ausencias inesperadas.
Adaptación (Integración y Ambiente Laboral):

Estabilización de Periodos Críticos: El modelo FAI busca identificar y mitigar los factores que causaban bajas históricas en meses como febrero, julio y diciembre. Al mejorar la integración del personal y el ambiente laboral, se reduce la incertidumbre y el impacto negativo en la producción durante estos meses. La meta es alinear estos meses con el rendimiento promedio del resto del año.
Resiliencia Operativa: Un equipo más cohesionado y adaptable es menos propenso a interrupciones significativas. La capacidad de la fuerza laboral para adaptarse a cambios en la demanda o en los procesos sin una caída drástica en la producción es clave.
Incentivo (Retención y Motivación del Empleado):

Reducción de la Rotación: Los programas de incentivos (más allá de la remuneración, incluyendo desarrollo profesional, reconocimiento y un ambiente de trabajo positivo) fomentan la lealtad y reducen la alta rotación de personal, que fue identificada como una de las principales causas de la variabilidad en 2025.
Aumento de la Moral y Compromiso: Un empleado motivado y reconocido tiende a ser más productivo y a buscar la excelencia en su trabajo. Esto impulsa una mejora continua en la calidad y cantidad de la producción.
Inversión en Capital Humano: Al invertir en el desarrollo y bienestar de los empleados, la empresa asegura un retorno en forma de mayor productividad y menor costo asociado a la constante contratación y capacitación de nuevo personal.
Conexión con la Proyección de Datos:

La técnica predictiva utilizada para 2026 (regresión lineal sobre la tendencia inicial positiva de 2026 y un factor de "elevación teórica" para los meses futuros) se basa en la expectativa de que el modelo FAI generará un crecimiento sostenido. El "factor de elevación teórica" (donde la producción proyectada para 2026 debe ser al menos un 15% más alta que el mismo mes en 2025 para los meses de mayo a diciembre) es una manera de cuantificar esta expectativa de mejora intrínseca derivada del programa de capacitación y retención.

En resumen, la proyección tan positiva se debe a la anticipación de que una fuerza laboral bien formada, adaptada y motivada, gracias al modelo FAI, no solo eliminará las anomalías negativas del pasado (como la caída de diciembre de 2025) sino que también impulsará la producción mensual a nuevos niveles de eficiencia y volumen a lo largo de todo el año.

```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression

# --- Global variables to store data ---
df_current = None
monthly_df_current = None
file_name_current = "Produccion_Planta_Z_GPO04_2025.csv" # Predefined file

# --- Utility functions ---
def load_and_process_data_from_path(filepath):
    global df_current, monthly_df_current, file_name_current

    try:
        temp_df = pd.read_csv(filepath)

        required_columns = ['dates', 'produccion']
        if not all(col in temp_df.columns for col in required_columns):
            missing_cols = [col for col in required_columns if col not in temp_df.columns]
            output_area.append_stdout(f"Error de Datos: El archivo CSV '{filepath}' no contiene las columnas requeridas: {', '.join(missing_cols)}.\n")
            reset_state()
            return False

        try:
            temp_df['dates'] = pd.to_datetime(temp_df['dates'], errors='raise')
        except Exception:
            try:
                temp_df['dates'] = pd.to_datetime(temp_df['dates'], format='%b %d', errors='raise')
            except Exception:
                output_area.append_stdout("Error de Fecha: No se pudo convertir la columna 'dates' a formato de fecha válido. Asegúrese de que sea 'YYYY-MM-DD' o 'Mon Day'.\n")
                reset_state()
                return False

        temp_df['month_num'] = temp_df['dates'].dt.month

        monthly_production_series = temp_df.groupby('month_num')['produccion'].sum().sort_index()

        spanish_month_names_list = [
            'Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
            'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'
        ]
        all_months_df = pd.DataFrame({'month_num': range(1, 13)})
        all_months_df['month_name'] = all_months_df['month_num'].apply(lambda x: spanish_month_names_list[x-1])

        monthly_df_current = pd.merge(all_months_df, monthly_production_series.reset_index(), on='month_num', how='left')
        monthly_df_current.rename(columns={'produccion': 'produccion'}, inplace=True)
        monthly_df_current['produccion'] = monthly_df_current['produccion'].fillna(0)

        df_current = temp_df
        output_area.append_stdout(f"Datos cargados y procesados correctamente desde '{filepath}'.\n")
        return True

    except Exception as e:
        output_area.append_stdout(f"Ocurrió un error al cargar o procesar los datos desde '{filepath}': {e}\n")
        reset_state()
        return False

def reset_state():
    global df_current, monthly_df_current, file_name_current
    df_current = None
    monthly_df_current = None
    # file_name_current remains predefined
    for btn in plot_buttons.values():
        btn.disabled = True
    output_area.clear_output()
    output_area.append_stdout(f"Error al cargar el archivo '{file_name_current}'.\n")

def enable_plot_buttons():
    for btn in plot_buttons.values():
        btn.disabled = False

def generate_plot(plot_type):
    if monthly_df_current is None:
        with output_area: clear_output(); output_area.append_stdout("Los datos no están cargados. Por favor, asegúrese de que el archivo CSV exista y tenga el formato correcto.\n")
        return

    df_plot = monthly_df_current.copy()

    with output_area:  # Redirect plot to the output widget
        clear_output(wait=True)
        plt.figure(figsize=(14, 7))

        if plot_type == 'bar':
            plt.bar(df_plot['month_name'], df_plot['produccion'], color='skyblue')
            plt.title(f'Producción Mensual: {file_name_current}', fontsize=24, fontweight='bold')
            plt.xlabel('Mes', fontsize=20, fontweight='bold')
            plt.ylabel('Producción Total', fontsize=20, fontweight='bold')
            plt.xticks(rotation=45, ha='right', fontsize=16, fontweight='bold')
            plt.yticks(fontsize=16, fontweight='bold')
            plt.grid(axis='y', linestyle='--', alpha=0.7)

        elif plot_type == 'line':
            plt.plot(df_plot['month_name'], df_plot['produccion'], color='red', marker='o', linestyle='-', linewidth=2)
            plt.title(f'Tendencia de Producción Mensual: {file_name_current}', fontsize=24, fontweight='bold')
            plt.xlabel('Mes', fontsize=20, fontweight='bold')
            plt.ylabel('Producción Total', fontsize=20, fontweight='bold')
            plt.xticks(rotation=45, ha='right', fontsize=16, fontweight='bold')
            plt.yticks(fontsize=16, fontweight='bold')
            plt.grid(axis='both', linestyle='--', alpha=0.7)

        elif plot_type == 'donut':
            df_plot_non_zero = df_plot[df_plot['produccion'] > 0]
            if df_plot_non_zero.empty:
                output_area.append_stdout("No hay producción para graficar en el gráfico de anillo.\n")
                plt.close() # Close empty figure
                return

            wedges, texts, autotexts = plt.pie(
                df_plot_non_zero['produccion'],
                labels=df_plot_non_zero['month_name'],
                autopct=lambda p: f'{p:.1f}%' if p > 1 else '',
                startangle=90,
                pctdistance=0.85,
                colors=plt.cm.Paired.colors
            )
            centre_circle = plt.Circle((0,0), 0.70, fc='white')
            plt.gca().add_artist(centre_circle)

            for text in texts:
                text.set_fontsize(14)
                text.set_fontweight('bold')
            for autotext in autotexts:
                autotext.set_fontsize(12)
                autotext.set_fontweight('bold')
                autotext.set_color('black')

            plt.title(f'Distribución Porcentual de Producción Mensual: {file_name_current}', fontsize=24, fontweight='bold', pad=20)
            plt.axis('equal')

        elif plot_type == 'regression':
            # Ensure enough data points for regression
            if len(df_plot['month_num'].unique()) < 2:
                output_area.append_stdout("No hay suficientes puntos de datos para generar una regresión lineal (se requieren al menos 2 meses con producción).\n")
                plt.close()
                return

            sns.regplot(x='month_num', y='produccion', data=df_plot, ci=95, line_kws={'color':'red'}, scatter_kws={'color':'blue'})
            plt.title(f'Regresión Lineal de Producción Mensual: {file_name_current}', fontsize=24, fontweight='bold')
            plt.xlabel('Mes (Número)', fontsize=20, fontweight='bold')
            plt.ylabel('Producción Total', fontsize=20, fontweight='bold')
            plt.xticks(ticks=df_plot['month_num'].unique(), labels=df_plot['month_name'], rotation=45, ha='right', fontsize=16, fontweight='bold')
            plt.yticks(fontsize=16, fontweight='bold')
            plt.grid(axis='both', linestyle='--', alpha=0.7)

        elif plot_type == 'histogram':
            # Using the full df_current for a more granular histogram if available, else monthly
            if df_current is not None and 'produccion' in df_current.columns:
                plt.hist(df_current['produccion'], bins=20, edgecolor='black', color='lightgreen')
                plt.title(f'Histograma de Frecuencia de Producción Diaria: {file_name_current}', fontsize=24, fontweight='bold')
                plt.xlabel('Cantidad de Producción Diaria', fontsize=20, fontweight='bold')
                plt.ylabel('Frecuencia', fontsize=20, fontweight='bold')
                plt.xticks(fontsize=16, fontweight='bold')
                plt.yticks(fontsize=16, fontweight='bold')
            else:
                # Fallback to monthly if daily data isn't suitable or available
                plt.hist(df_plot['produccion'], bins=min(len(df_plot), 10), edgecolor='black', color='lightgreen')
                plt.title(f'Histograma de Frecuencia de Producción Mensual: {file_name_current}', fontsize=24, fontweight='bold')
                plt.xlabel('Cantidad de Producción Mensual', fontsize=20, fontweight='bold')
                plt.ylabel('Frecuencia', fontsize=20, fontweight='bold')
                plt.xticks(fontsize=16, fontweight='bold')
                plt.yticks(fontsize=16, fontweight='bold')

        plt.tight_layout()
        plt.show()

# --- ipywidgets Layout ---

# Buttons for plot generation
plot_buttons = {
    'bar': widgets.Button(description='Gráfica de Barras', button_style='primary', disabled=True, layout=widgets.Layout(width='auto', margin='5px 0')),
    'line': widgets.Button(description='Gráfica de Líneas', button_style='primary', disabled=True, layout=widgets.Layout(width='auto', margin='5px 0')),
    'donut': widgets.Button(description='Gráfica de Anillo', button_style='primary', disabled=True, layout=widgets.Layout(width='auto', margin='5px 0')),
    'regression': widgets.Button(description='Gráfica de Regresión Lineal', button_style='primary', disabled=True, layout=widgets.Layout(width='auto', margin='5px 0')),
    'histogram': widgets.Button(description='Histograma de Frecuencias', button_style='primary', disabled=True, layout=widgets.Layout(width='auto', margin='5px 0'))
}

plot_buttons['bar'].on_click(lambda x: generate_plot('bar'))
plot_buttons['line'].on_click(lambda x: generate_plot('line'))
plot_buttons['donut'].on_click(lambda x: generate_plot('donut'))
plot_buttons['regression'].on_click(lambda x: generate_plot('regression'))
plot_buttons['histogram'].on_click(lambda x: generate_plot('histogram'))

# Output area for messages and plots
output_area = widgets.Output()

# Combine widgets into a layout
controls_layout = widgets.VBox([
    widgets.HTML("<h3>Dashboard Interactivo de Análisis de Producción</h3>"),
    widgets.Label("Archivo de datos cargado: Produccion_Planta_Z_GPO04_2025.csv"), # Info for the user
    widgets.Label("--- Generar Gráficas ---"),
    plot_buttons['bar'],
    plot_buttons['line'],
    plot_buttons['donut'],
    plot_buttons['regression'],
    plot_buttons['histogram'],
    widgets.Label("\n--- Resultados y Gráficas ---")
], layout=widgets.Layout(width='300px', padding='10px', border='2px solid lightgray'))

display(widgets.HBox([controls_layout, output_area]))

# Initial data load
with output_area:
    clear_output(wait=True)
    if load_and_process_data_from_path('/content/Proyecto_Zihumara/Produccion_Planta_Z_GPO04_2025.csv'):
        enable_plot_buttons()
```

```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression

# --- Global variables for this 2026 dashboard ---
df_2026_dashboard = None
monthly_df_2026_dashboard = None
file_name_2026_dashboard = "/content/Proyecto_Zihumara/Produccion_2026_troqueles.csv" # Predefined file
projected_monthly_production_2026 = {} # New global variable for storing projected monthly values

# --- Utility functions ---
def load_and_process_data_2026_dashboard():
    global df_2026_dashboard, monthly_df_2026_dashboard

    filepath = file_name_2026_dashboard
    output_area_2026.append_stdout(f"Intentando cargar y procesar datos desde '{filepath}'\n")

    try:
        temp_df = pd.read_csv(filepath)

        required_columns = ['dates', 'produccion']
        if not all(col in temp_df.columns for col in required_columns):
            missing_cols = [col for col in required_columns if col not in temp_df.columns]
            output_area_2026.append_stdout(f"Error de Datos: El archivo CSV '{filepath}' no contiene las columnas requeridas: {', '.join(missing_cols)}.\n")
            reset_state_2026()
            return False

        try:
            temp_df['dates'] = pd.to_datetime(temp_df['dates'], errors='raise')
        except Exception:
            try:
                temp_df['dates'] = pd.to_datetime(temp_df['dates'], format='%b %d', errors='raise')
            except Exception:
                output_area_2026.append_stdout("Error de Fecha: No se pudo convertir la columna 'dates' a formato de fecha válido. Asegúrese de que sea 'YYYY-MM-DD' o 'Mon Day'.\n")
                reset_state_2026()
                return False

        temp_df['month_num'] = temp_df['dates'].dt.month

        monthly_production_series = temp_df.groupby('month_num')['produccion'].sum().sort_index()

        spanish_month_names_list = [
            'Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
            'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'
        ]
        all_months_df = pd.DataFrame({'month_num': range(1, 13)}) # All 12 months for consistency
        all_months_df['month_name'] = all_months_df['month_num'].apply(lambda x: spanish_month_names_list[x-1])

        # Merge with actual production data
        monthly_df_2026_dashboard = pd.merge(all_months_df, monthly_production_series.reset_index(), on='month_num', how='left')
        monthly_df_2026_dashboard.rename(columns={'produccion': 'produccion'}, inplace=True)
        monthly_df_2026_dashboard['produccion'] = monthly_df_2026_dashboard['produccion'].fillna(0) # Fill months without data with 0

        df_2026_dashboard = temp_df # Store the full DataFrame too
        output_area_2026.append_stdout(f"Datos cargados y procesados correctamente desde '{filepath}'.\n")
        return True

    except FileNotFoundError:
        output_area_2026.append_stdout(f"Error: El archivo '{filepath}' no fue encontrado. Asegúrese de que la ruta sea correcta.\n")
        reset_state_2026()
        return False
    except Exception as e:
        output_area_2026.append_stdout(f"Ocurrió un error al cargar o procesar los datos desde '{filepath}': {e}\n")
        reset_state_2026()
        return False

def reset_state_2026():
    global df_2026_dashboard, monthly_df_2026_dashboard
    df_2026_dashboard = None
    monthly_df_2026_dashboard = None
    for btn in plot_buttons_2026.values():
        btn.disabled = True
    output_area_2026.clear_output()
    output_area_2026.append_stdout(f"Error al cargar el archivo '{file_name_2026_dashboard}'.\n")

def enable_plot_buttons_2026():
    for btn in plot_buttons_2026.values():
        btn.disabled = False

def generate_plot_2026(plot_type):
    if monthly_df_2026_dashboard is None:
        with output_area_2026: clear_output(); output_area_2026.append_stdout("Los datos no están cargados. Por favor, asegúrese de que el archivo CSV exista y tenga el formato correcto.\n")
        return

    df_plot = monthly_df_2026_dashboard.copy()

    with output_area_2026:  # Redirect plot to the output widget
        clear_output(wait=True)
        plt.figure(figsize=(14, 7))

        if plot_type == 'bar':
            # Filter to show only months with actual data
            actual_months_df = df_plot[df_plot['produccion'] > 0]
            if actual_months_df.empty:
                output_area_2026.append_stdout("No hay datos de producción para graficar en el gráfico de barras.\n")
                plt.close()
                return

            plt.bar(actual_months_df['month_name'], actual_months_df['produccion'], color='skyblue')
            plt.title(f'Producción Mensual 2026 (Actual): {file_name_2026_dashboard}', fontsize=24, fontweight='bold')
            plt.xlabel('Mes', fontsize=20, fontweight='bold')
            plt.ylabel('Producción Total', fontsize=20, fontweight='bold')
            plt.xticks(rotation=45, ha='right', fontsize=16, fontweight='bold')
            plt.yticks(fontsize=16, fontweight='bold')
            plt.grid(axis='y', linestyle='--', alpha=0.7)

        elif plot_type == 'line':
            actual_months_df = df_plot[df_plot['produccion'] > 0]
            if actual_months_df.empty:
                output_area_2026.append_stdout("No hay datos de producción para graficar en el gráfico de líneas.\n")
                plt.close()
                return

            plt.plot(actual_months_df['month_name'], actual_months_df['produccion'], color='red', marker='o', linestyle='-', linewidth=2)
            plt.title(f'Tendencia de Producción Mensual 2026 (Actual): {file_name_2026_dashboard}', fontsize=24, fontweight='bold')
            plt.xlabel('Mes', fontsize=20, fontweight='bold')
            plt.ylabel('Producción Total', fontsize=20, fontweight='bold')

            plt.xticks(rotation=45, ha='right', fontsize=16, fontweight='bold')
            plt.yticks(fontsize=16, fontweight='bold')

            plt.grid(axis='both', linestyle='--', alpha=0.7)

        elif plot_type == 'donut':
            df_plot_non_zero = df_plot[df_plot['produccion'] > 0]
            if df_plot_non_zero.empty:
                output_area_2026.append_stdout("No hay producción para graficar en el gráfico de anillo.\n")
                plt.close() # Close empty figure
                return

            wedges, texts, autotexts = plt.pie(
                df_plot_non_zero['produccion'],
                labels=df_plot_non_zero['month_name'],
                autopct=lambda p: f'{p:.1f}%' if p > 1 else '',
                startangle=90,
                pctdistance=0.85,
                colors=plt.cm.Paired.colors
            )
            centre_circle = plt.Circle((0,0), 0.70, fc='white')
            plt.gca().add_artist(centre_circle)

            for text in texts:
                text.set_fontsize(14)
                text.set_fontweight('bold')
            for autotext in autotexts:
                autotext.set_fontsize(12)
                autotext.set_fontweight('bold')
                autotext.set_color('black')

            plt.title(f'Distribución Porcentual de Producción Mensual 2026 (Actual): {file_name_2026_dashboard}', fontsize=24, fontweight='bold', pad=20)
            plt.axis('equal')

        elif plot_type == 'regression':
            actual_months_df = df_plot[df_plot['produccion'] > 0]
            if len(actual_months_df['month_num'].unique()) < 2:
                output_area_2026.append_stdout("No hay suficientes puntos de datos para generar una regresión lineal (se requieren al menos 2 meses con producción).\n")
                plt.close()
                return

            sns.regplot(x='month_num', y='produccion', data=actual_months_df, ci=95, line_kws={'color':'red'}, scatter_kws={'color':'blue'})
            plt.title(f'Regresión Lineal de Producción Mensual 2026 (Actual): {file_name_2026_dashboard}', fontsize=24, fontweight='bold')
            plt.xlabel('Mes (Número)', fontsize=20, fontweight='bold')
            plt.ylabel('Producción Total', fontsize=20, fontweight='bold')
            plt.xticks(ticks=actual_months_df['month_num'].unique(), labels=actual_months_df['month_name'], rotation=45, ha='right', fontsize=16, fontweight='bold')
            plt.yticks(fontsize=16, fontweight='bold')
            plt.grid(axis='both', linestyle='--', alpha=0.7)

        elif plot_type == 'histogram':
            # Use the full df_2026_dashboard for a more granular histogram if available, else monthly
            if df_2026_dashboard is not None and 'produccion' in df_2026_dashboard.columns:
                plt.hist(df_2026_dashboard['produccion'], bins=20, edgecolor='black', color='lightgreen')
                plt.title(f'Histograma de Frecuencia de Producción Diaria 2026 (Actual): {file_name_2026_dashboard}', fontsize=24, fontweight='bold')
                plt.xlabel('Cantidad de Producción Diaria', fontsize=20, fontweight='bold')
                plt.ylabel('Frecuencia', fontsize=20, fontweight='bold')
                plt.xticks(fontsize=16, fontweight='bold')
                plt.yticks(fontsize=16, fontweight='bold')
            else:
                actual_months_df = df_plot[df_plot['produccion'] > 0]
                if actual_months_df.empty:
                    output_area_2026.append_stdout("No hay datos de producción para graficar el histograma.\n")
                    plt.close()
                    return
                plt.hist(actual_months_df['produccion'], bins=min(len(actual_months_df), 10), edgecolor='black', color='lightgreen')
                plt.title(f'Histograma de Frecuencia de Producción Mensual 2026 (Actual): {file_name_2026_dashboard}', fontsize=24, fontweight='bold')
                plt.xlabel('Cantidad de Producción Mensual', fontsize=20, fontweight='bold')
                plt.ylabel('Frecuencia', fontsize=20, fontweight='bold')
                plt.xticks(fontsize=16, fontweight='bold')
                plt.yticks(fontsize=16, fontweight='bold')

        plt.tight_layout()
        plt.show()

def _get_full_projection_df_2026():
    """Helper function to calculate the full year's actual + projected production."""
    global projected_monthly_production_2026 # Declare global

    if monthly_df_2026_dashboard is None:
        return None

    actual_data_2026 = monthly_df_2026_dashboard[monthly_df_2026_dashboard['produccion'] > 0].copy()

    if len(actual_data_2026) < 2:
        return None

    X_train = actual_data_2026['month_num'].values.reshape(-1, 1)
    y_train = actual_data_2026['produccion'].values

    model = LinearRegression()
    model.fit(X_train, y_train)

    X_predict_all_months = np.array(range(1, 13)).reshape(-1, 1)
    y_predict_all_months = model.predict(X_predict_all_months)
    y_predict_all_months[y_predict_all_months < 0] = 0

    projected_full_year_df = pd.DataFrame({
        'month_num': range(1, 13),
        'produccion': y_predict_all_months.astype(int)
    })
    spanish_month_names_list = [
        'Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
        'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'
    ]
    projected_full_year_df['month_name'] = projected_full_year_df['month_num'].apply(lambda x: spanish_month_names_list[x-1])

    final_plot_df = projected_full_year_df.copy()
    for index, row in actual_data_2026.iterrows():
        final_plot_df.loc[final_plot_df['month_num'] == row['month_num'], 'produccion'] = row['produccion']

    # Store projected values in global variables
    projected_monthly_production_2026 = {} # Initialize the dictionary
    for _, row in final_plot_df.iterrows():
        month_spanish_name = spanish_month_names_list[int(row['month_num']) - 1]
        var_name = f'ProduccionProyectada{month_spanish_name}26'
        globals()[var_name] = int(row['produccion']) # Store as integer in individual global variables
        projected_monthly_production_2026[month_spanish_name] = int(row['produccion']) # Also store in dictionary

    return final_plot_df, actual_data_2026

def generate_projection_plot_2026():
    projection_result = _get_full_projection_df_2026()
    if projection_result is None:
        with output_area_2026: clear_output(); output_area_2026.append_stdout("No hay suficientes datos para generar la proyección completa.\n")
        return

    final_plot_df, actual_data_2026 = projection_result
    spanish_month_names_list = [
        'Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
        'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'
    ]

    with output_area_2026:
        clear_output(wait=True)

        fig, axes = plt.subplots(2, 1, figsize=(16, 16)) # Two rows, one column

        # --- First Subplot (Existing Bar Chart with Regression Line) ---
        ax1 = axes[0] # First subplot

        # Plot actual data with a specific color
        ax1.bar(actual_data_2026['month_name'], actual_data_2026['produccion'], color='skyblue', label='Producción Real 2026', width=0.4)

        # Plot projected data for future months (beyond actual data) with a different color
        predicted_future_months_df = final_plot_df[final_plot_df['month_num'] > actual_data_2026['month_num'].max()]
        if not predicted_future_months_df.empty:
            ax1.bar(predicted_future_months_df['month_name'], predicted_future_months_df['produccion'], color='lightcoral', label='Producción Proyectada 2026', width=0.4)

        # Plot the regression line over all 12 months for visual trend
        # This line uses the 'produccion' from projected_full_year_df which is the raw linear regression fit
        ax1.plot(final_plot_df['month_name'], final_plot_df['produccion'], color='green', linestyle='--', marker='x', label='Línea de Tendencia (Regresión)', linewidth=2)


        ax1.set_title(f'Producción Mensual 2026: Real y Proyectada ({file_name_2026_dashboard})', fontsize=20, fontweight='bold')
        ax1.set_xlabel('Mes', fontsize=16, fontweight='bold')
        ax1.set_ylabel('Producción Total', fontsize=16, fontweight='bold')
        ax1.set_xticks(range(len(spanish_month_names_list))) # Set x-ticks for all 12 months
        ax1.set_xticklabels(spanish_month_names_list, rotation=45, ha='right', fontsize=14, fontweight='bold')
        ax1.tick_params(axis='y', labelsize=14)
        ax1.legend(fontsize=12)
        ax1.grid(axis='y', linestyle='--', alpha=0.7)

        # --- Second Subplot (New Line Chart: Actual vs. Full Projection) ---
        ax2 = axes[1] # Second subplot

        # Actual production (red line)
        ax2.plot(actual_data_2026['month_name'], actual_data_2026['produccion'], color='red', marker='o', linestyle='-', linewidth=2, label='Producción Real (Hasta el último mes disponible)')

        # Full year combined production (green line)
        ax2.plot(final_plot_df['month_name'], final_plot_df['produccion'], color='green', marker='x', linestyle='--', linewidth=2, label='Producción Total (Real + Proyectada)')

        ax2.set_title(f'Tendencia de Producción 2026: Real vs. Total Proyectada', fontsize=20, fontweight='bold')
        ax2.set_xlabel('Mes', fontsize=16, fontweight='bold')
        ax2.set_ylabel('Producción Total', fontsize=16, fontweight='bold')
        ax2.set_xticks(range(len(spanish_month_names_list))) # Set x-ticks for all 12 months
        ax2.set_xticklabels(spanish_month_names_list, rotation=45, ha='right', fontsize=14, fontweight='bold')
        ax2.tick_params(axis='y', labelsize=14)
        ax2.legend(fontsize=12)
        ax2.grid(axis='both', linestyle='--', alpha=0.7)

        plt.tight_layout()
        display(fig) # Use display(fig) instead of plt.show()
        plt.close(fig) # Close the figure to free up memory

        # Removed printing of total_projected_annual and monthly details

def generate_projected_line_plot_2026():
    projection_result = _get_full_projection_df_2026()
    if projection_result is None:
        with output_area_2026: clear_output(); output_area_2026.append_stdout("No hay suficientes datos para generar el gráfico de líneas proyectado.\n")
        return

    final_plot_df, actual_data_2026 = projection_result
    spanish_month_names_list = [
        'Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
        'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'
    ]

    with output_area_2026:
        clear_output(wait=True)
        fig, ax = plt.subplots(figsize=(14, 7)) # Create figure and axes explicitly

        # Plot the full year's combined actual and projected production (green line)
        ax.plot(final_plot_df['month_name'], final_plot_df['produccion'], color='green', marker='o', linestyle='-', linewidth=2, label='Producción Total (Real + Proyectada)')

        ax.set_title(f'Gráfico de Líneas: Producción 2026 Total Proyectada', fontsize=24, fontweight='bold')
        ax.set_xlabel('Mes', fontsize=20, fontweight='bold')
        ax.set_ylabel('Producción Total', fontsize=20, fontweight='bold')
        ax.set_xticks(range(len(spanish_month_names_list)))
        ax.set_xticklabels(spanish_month_names_list, rotation=45, ha='right', fontsize=16, fontweight='bold')
        ax.tick_params(axis='y', labelsize=16) # Use ax.tick_params for y-axis
        ax.legend(fontsize=14)
        ax.grid(axis='both', linestyle='--', alpha=0.7)
        fig.tight_layout() # Use fig.tight_layout()

        display(fig) # Explicitly display the figure here

        # Removed printing of total_projected_annual and monthly details
        plt.close(fig) # Close the figure to free up memory


# --- ipywidgets Layout ---

# Buttons for plot generation for 2026 data
plot_buttons_2026 = {
    'bar': widgets.Button(description='Gráfica de Barras (Actual)', button_style='primary', disabled=True, layout=widgets.Layout(width='auto', margin='5px 0')),
    'line': widgets.Button(description='Gráfica de Líneas (Actual)', button_style='primary', disabled=True, layout=widgets.Layout(width='auto', margin='5px 0')),
    'donut': widgets.Button(description='Gráfica de Anillo (Actual)', button_style='primary', disabled=True, layout=widgets.Layout(width='auto', margin='5px 0')),
    'regression': widgets.Button(description='Regresión Lineal (Actual)', button_style='primary', disabled=True, layout=widgets.Layout(width='auto', margin='5px 0')),
    'histogram': widgets.Button(description='Histograma (Actual)', button_style='primary', disabled=True, layout=widgets.Layout(width='auto', margin='5px 0')),
    'projection': widgets.Button(description='Proyección de Producción (Completa)', button_style='success', disabled=True, layout=widgets.Layout(width='auto', margin='5px 0')),
    'projected_line': widgets.Button(description='Gráfica de Líneas Proyectada', button_style='success', disabled=True, layout=widgets.Layout(width='auto', margin='5px 0'))
}

plot_buttons_2026['bar'].on_click(lambda x: generate_plot_2026('bar'))
plot_buttons_2026['line'].on_click(lambda x: generate_plot_2026('line'))
plot_buttons_2026['donut'].on_click(lambda x: generate_plot_2026('donut'))
plot_buttons_2026['regression'].on_click(lambda x: generate_plot_2026('regression'))
plot_buttons_2026['histogram'].on_click(lambda x: generate_plot_2026('histogram'))
plot_buttons_2026['projection'].on_click(lambda x: generate_projection_plot_2026())
plot_buttons_2026['projected_line'].on_click(lambda x: generate_projected_line_plot_2026())

# Output area for messages and plots
output_area_2026 = widgets.Output()

# Combine widgets into a layout
controls_layout_2026 = widgets.VBox([
    widgets.HTML("<h3>Dashboard Interactivo de Análisis de Producción 2026</h3>"),
    widgets.Label(f"Archivo de datos cargado: {file_name_2026_dashboard}"),
    widgets.Label("--- Generar Gráficas (Datos Actuales) ---"),
    plot_buttons_2026['bar'],
    plot_buttons_2026['line'],
    plot_buttons_2026['donut'],
    plot_buttons_2026['regression'],
    plot_buttons_2026['histogram'],
    widgets.Label("--- Proyección Completa (hasta Diciembre) ---"),
    plot_buttons_2026['projection'],
    plot_buttons_2026['projected_line'], # New button added here
    widgets.Label("\n--- Resultados y Gráficas ---")
], layout=widgets.Layout(width='350px', padding='10px', border='2px solid lightgray'))

display(widgets.HBox([controls_layout_2026, output_area_2026]))

# Initial data load
with output_area_2026:
    clear_output(wait=True)
    if load_and_process_data_2026_dashboard():
        enable_plot_buttons_2026()
    else:
        output_area_2026.append_stdout("No se pudieron cargar los datos iniciales. Verifique la ruta del archivo y su formato.\n")
```
